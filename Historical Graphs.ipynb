```python
ncfp = pd.read_csv("NCFP.csv", 
                    usecols=["lastupdate", "metricname", "ptsseen_avg"],
                    parse_dates=["lastupdate"])
pivot = ncfp.pivot_table(index='lastupdate', 
                    columns='metricname', 
                    values="ptsseen_avg")
pivot
```

```python
pivot["COLORECTAL"].plot()
```
